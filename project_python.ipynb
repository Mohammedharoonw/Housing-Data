6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.


```python
# we made a function name calculate_median for getting the median fro required data from the dataset
def calculate_median(column_name):
    housing_data = pd.read_excel("housing.xlsx")
    # We get the median value with median function
    median_value = housing_data[column_name].median()
    # return function is used to return the calculated median value 
    return median_value

median_value = calculate_median("population")
print(median_value)

```

**This function can be used to calculate the median of any column in the housing dataset  by passing the column name as an argument to the calculate_median() function.**

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# define a function to categorize total bedroom size
def size_of_bedroom(total_bedrooms):
    if total_bedrooms <= 10:
        return "small"
    elif total_bedrooms < 1000:
        return "medium"
    else:
        return "large"

# Apply to create the new column
housing_data["total_bedroom_size"] = housing_data["total_bedrooms"].apply(size_of_bedroom)

# Print the first 5 rows of the column
print(housing_data.head())
```

**In this we have created a new folder for bedroom sizes by putting the condition to small medium large now in the data new column is been made name bedroom size**
